```python
import numpy as np
import pandas as pd
```

```python
submission_1 = pd.read_csv('./data/Model_1_CatBoost.csv')
submission_2 = pd.read_csv('./data/Model_2_CatBoost.csv')
submission_3 = pd.read_csv('./data/Model_3_CatBoost.csv')
submission_4 = pd.read_csv('./data/Model_4_CatBoost.csv')
submission_5 = pd.read_csv('./data/Model_5_CatBoost.csv')
submission_6 = pd.read_csv('./data/Model_6_CatBoost.csv')
```

```python
submission_a = pd.merge(submission_1, submission_3, how = 'left', on = 'ParcelId')
submission_a.head()
```

```python
submission_b = pd.merge(submission_a, submission_5, how = 'left', on = 'ParcelId')
submission_b.head()
```

```python
submission_c = pd.merge(submission_b, submission_2, how = 'left', on = 'ParcelId')
submission_c.head()
```

```python
submission_d = pd.merge(submission_c, submission_4, how = 'left', on = 'ParcelId')
submission_d.head()
```

```python
submission = pd.merge(submission_d, submission_6, how = 'left', on = 'ParcelId')
submission.head()
```

```python
submission.to_csv('./data/submission.csv', index = False)
```
